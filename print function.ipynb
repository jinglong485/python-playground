```python
import threading
from queue import Queue
from general import *
```

```python
NUMBER_OF_THREADS = 8
queue = Queue()
def create_workers():
    for _ in range(NUMBER_OF_THREADS):
        t = threading.Thread(target = work, daemon = True)
        t.start()
```

```python
def work():
    while True:
        line = queue.get()
        print(line)
        print('The threading now using is: {}'.format(threading.current_thread().getName()))
        queue.task_done()
```

```python
def go():
    queued_links = file_to_list('metart2017.list')
    if len(queued_links) > 0:
        for link in file_to_list('metart2017.list'):
            queue.put(link)
    queue.join()
```

```python
create_workers()
go()
```
